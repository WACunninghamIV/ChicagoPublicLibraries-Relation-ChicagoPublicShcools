```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Libararies_Chicago.csv"
library_df = pd.read_csv(csv_file)
library_df.head()
```

```python
library_df.reset_index(inplace = True)
library_df.columns = library_df.iloc[0]
library_df.drop(library_df.index[0],inplace=True)
library_df.head()
```

```python
csv_file = "Public_Schools_Chicago.csv"
school_df = pd.read_csv(csv_file)
school_df.head()
```

```python
school_df.reset_index(inplace = True)
school_df.columns = school_df.iloc[0]
school_df.drop(school_df.index[0],inplace=True)
school_df.head()
```

```python
school_df.rename(index = {"ZIP Code":"ZIP"}, 
          inplace = True)
```

```python
school_df
```

```python
lib = pd.read_csv("Libararies_Chicago.csv")
public = pd.read_csv("Public_Schools_Chicago.csv")

merged = lib.merge(public, on='ZIP')

merged.to_csv("Result.csv", index=False)
```
